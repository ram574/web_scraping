```python
final_raw_data= []
dates = []
pck_names = []
descrips = []
links = []

for page_number in range(1,11):
    file_html = open('Data/Html_Data_test/{}.html'.format(page_number), 'rb')
    plain_text_1 = file_html.read()
    soup_1 = BeautifulSoup(plain_text_1, "lxml")
    
    for date_1 in soup_1.find_all('time', {'data-controller':"localized-time"}):
        dates.append(date_1.getText())
        
    for pck_name_1 in soup_1.find_all('span', {'class':"package-snippet__name"}):
        pck_names.append(pck_name_1.getText())
        
    for descrips_1 in soup_1.find_all('p', {'class': "package-snippet__description"}):
        descrips.append(descrips_1.getText())
        
    for link_1 in soup_1.find_all('a', {'class':"package-snippet"}):
        links.append('https://pypi.org/' + link_1.get('href'))
```
